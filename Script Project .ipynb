#### Checking outliers

```python
# function to separate outlier or not
def check_outlier(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    return ['outlier' if (x<q1-1.5*iqr or x>q3+1.5*iqr) 
                                 else 'not-outlier' for x in data[column].values]
```

```python
df_outlier = df.copy() # copy data set
```

```python
df_outlier['height_outlier'] = check_outlier(data = df, column = 'Height_(cm)')
df_outlier['weight_outlier'] = check_outlier(data = df, column = 'Weight_(kg)')
df_outlier['BMI_outlier'] = check_outlier(data = df, column = 'BMI')
df_outlier['alcohol_outlier'] = check_outlier(data = df, column = 'Alcohol_Consumption')
df_outlier['fruit_outlier'] = check_outlier(data = df, column = 'Fruit_Consumption')
df_outlier['vegetables_outlier'] = check_outlier(data = df, column = 'Green_Vegetables_Consumption')
df_outlier['potato_outlier'] = check_outlier(data = df, column = 'FriedPotato_Consumption')
```

```python
df_outlier = df_outlier[(df_outlier['height_outlier'] != 'outlier') & (df_outlier['weight_outlier'] != 'outlier') & (df_outlier['BMI_outlier'] != 'outlier') & (df_outlier['alcohol_outlier'] != 'outlier') & (df_outlier['fruit_outlier'] != 'outlier') & (df_outlier['vegetables_outlier'] != 'outlier') & (df_outlier['potato_outlier'] != 'outlier')]
df_outlier = df_outlier.drop(['height_outlier', 'weight_outlier', 'BMI_outlier', 'alcohol_outlier', 'fruit_outlier', 'vegetables_outlier', 'potato_outlier'], axis = 1)
```

# Data Preprocesing Without Outliers

## Encoding feature and target

```python
# encoding categorical column
map_yes_no = {
    "Yes" : 1,
    "No" : 0
}

map_sex = {
    "Female" : 1,
    "Male" : 0
}

map_diabetes = {
    "Yes" : 1,
    "No" : 0,
    "No, pre-diabetes or borderline diabetes" : 0,
    "Yes, but female told only during pregnancy" : 1
}

map_age = {
    "18-24" : 0,
    "25-29" : 1,
    "30-34" : 2,
    "35-39" : 3,
    "40-44" : 4,
    "45-49" : 5,
    "50-54" : 6,
    "55-59" : 7,
    "60-64" : 8,
    "65-69" : 9,
    "70-74" : 10,
    "75-79" : 11,
    "80+" : 12
}


map_general_health = {
    "Very Good" : 0,
    "Good" : 1,
    "Excellent" : 2,
    "Fair" : 3,
    "Poor" : 4 
}


map_chekup = {
    "Within the past year" : 0,
    "Within the past 2 years" : 1,
    "Within the past 5 years" : 2,
    "5 or more years ago" : 3,
    "Never" : 4
}

map_body = {
    "Under Weight" : 0,
    "Normal" : 1,
    "Over Weight" : 2,
    "Obesity" : 3
}
```

```python
# convert categorical into numerical
df_outlier['Heart_Disease'] = df_outlier["Heart_Disease"].map(map_yes_no)
df_outlier['Exercise'] = df_outlier["Exercise"].map(map_yes_no)
df_outlier['Skin_Cancer'] = df_outlier["Skin_Cancer"].map(map_yes_no)
df_outlier['Other_Cancer'] = df_outlier["Other_Cancer"].map(map_yes_no)
df_outlier['Depression'] = df_outlier["Depression"].map(map_yes_no)
df_outlier['Arthritis'] = df_outlier["Arthritis"].map(map_yes_no)
df_outlier['Smoking_History'] = df_outlier["Smoking_History"].map(map_yes_no)
df_outlier["General_Health"] = df_outlier['General_Health'].map(map_general_health)
df_outlier['Checkup'] = df_outlier['Checkup'].map(map_chekup)
df_outlier['Sex'] = df_outlier['Sex'].map(map_sex)
df_outlier['Diabetes'] = df_outlier['Diabetes'].map(map_diabetes)
df_outlier['Age_Category'] = df_outlier['Age_Category'].map(map_age)
```

## Train, Test and Split

```python
y = df_outlier['Heart_Disease']
X = df_outlier.drop("Heart_Disease", axis = 1)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test,y_train,y_test = train_test_split(X,
                                                y,
                                                test_size = 0.1,
                                                random_state = 42)
```

```python
y_test.value_counts()
```

```python
X_test.info()
```

i split the data into 10 % test, 90 % train. Than undersampling train data

```python
from imblearn import under_sampling
X_under, y_under = under_sampling.RandomUnderSampler(sampling_strategy = 0.8).fit_resample(X_train, y_train) # 80 % minority
```

```python
y_under.value_counts()
```

# Modeling with Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

xgd = GradientBoostingClassifier()


from sklearn.model_selection import RandomizedSearchCV
parameters = {
    'n_estimators' : (10,20,30,40,50),
    'max_depth' : (3,4,5)
}
xgd_clf_randomcv = RandomizedSearchCV(xgd, parameters, cv=5, 
                                     scoring='recall', n_iter=10)
xgd_clf_randomcv.fit(X_under,y_under) # train model
```

```python
cv_result = pd.DataFrame(xgd_clf_randomcv.cv_results_)
retain_cols = ['params','mean_test_score','rank_test_score']
cv_result[retain_cols].sort_values('rank_test_score')
```

```python
from sklearn.metrics import roc_auc_score, classification_report, f1_score
y_pred_train = xgd_clf_randomcv.best_estimator_.predict(X_under)
print(classification_report(y_under,y_pred_train))
```

```python
y_pred_test = xgd_clf_randomcv.best_estimator_.predict(X_test)
print(classification_report(y_test,y_pred_test))
```

interpretate  model :
1. best model is on paramter n_estimators : 30 and max_depth : 5
2. precision metric has overfitting, where the training model higher than testing
3. recall is a metric we focus on. From all occurrence of heart disease, the model caught 78 %
4. precision metric 23 % means that from all true positive predictions made by the model. only 23 % that true positive.


## Confusion Metric

```python
# plotting confusion matrix
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

fig, ax = plt.subplots(figsize=(8,5))
y_pred = xgd_clf_randomcv.best_estimator_.predict(X_test)

cm = confusion_matrix(y_test, y_pred, labels=xgd_clf_randomcv.best_estimator_.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=xgd_clf_randomcv.best_estimator_.classes_)
disp.plot(ax=ax)
plt.show()
```

observation of confusion metric

5449 people predicted as true positive CVD by model, or 29 % from the data set. On the other side, we lost 0.2 % from data test people that truth positive CVD, because the model predicted them as negatif.
